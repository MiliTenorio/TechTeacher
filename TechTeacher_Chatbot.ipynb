Chatbot configuration

```python
from google import genai
from google.genai import types

client = genai.Client()
model = "gemini-2.0-flash"

chat = client.chats.create(model=model)

chat_config = types.GenerateContentConfig(
    system_instruction="""
    You're a TechTeacher! A chatbot that helps people excited about technology
    who want to ask you questions about studying computer science and programming.
    But these people have never had any contact with computer science or programming.
    You always ask questions to find out the person's level of knowledge before answering.
    And your answers are simple with good practical examples to explain different technology concepts.
    """

)

chat = client.chats.create(model=model, config=chat_config)
```

Chatbot conversation

```python
response = chat.send_message("""Make an introduction about yourself and what you
can do for those who talk to you! And say that to finish you need to say ‘Bye’""")
print("TechTeacher: ", response.text)

prompt = input("Your prompt: ")

while prompt != "Bye":
  response = chat.send_message(prompt)
  print("TechTeacher: ", response.text)
  prompt = input("Your prompt: ")

response = chat.send_message("""Say goodbye to the person you were talking to
and give them some advice on the topics discussed in the conversation""")
print("TechTeacher: ", response.text)
```
